```python
# in categorical ones since they are not measured by numbers we can differentiate them as discrete.
```

```python
# and about the customer_id column we see that every each raws are unique. one specific for each
```

```python
def differentiate(x):

    continuous = []
    discrete = []
    
    for c in x:
        if len(x[c].unique()) < 10:
            discrete.append(c)
        else:
            continuous.append(c)
    return continuous, discrete

continuous, discrete = differentiate(numericals)
```

```python
discrete_df = pd.DataFrame(data = customer_df, columns=discrete)
discrete_df.head()
```

```python
continuous_df = pd.DataFrame(data = customer_df, columns=continuous)
display(continuous_df.shape)
continuous_df.head()
```

```python
correlations_matrix = continuous_df.corr()
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

```python
def plt_df(x):
    for i in x:
        sns.displot(discrete_df[i])
        plt.show()
plt_df(discrete_df)
```

```python
def plt_df(x):
    for i in x:
        sns.boxplot(continuous_df[i])
        plt.show()
plt_df(continuous_df)
```

```python
# From the plots it's seen that 'customer_lifetime_value', 'monthly_premium_auto' and 'total_claim_amount' have significantly outliers.
# If we transform it won't be for the final user. so we're going to bin the data. 
```

```python
def log_transform_continuous(x):
    if np.isfinite(x) and x!=0:
        return np.log(x)
    else:
        return np.NAN
```

```python
pd.Series(continuous_df['customer_lifetime_value'].apply(log_transform_continuous)).hist()
plt.show()
```

```python
continuous_df['customer_lifetime_value'] = continuous_df['customer_lifetime_value'].apply(log_transform_continuous)
```

```python
iqr = np.percentile(continuous_df['customer_lifetime_value'],75) - np.percentile(continuous_df['customer_lifetime_value'],25)
upper_limit = np.percentile(continuous_df['customer_lifetime_value'],75) + 1.5*iqr
print(upper_limit)

outliers = continuous_df[continuous_df['customer_lifetime_value'] > upper_limit]
len(outliers)
```

```python
# number of outliers seem not so significant. so they can be removed to be able to get rid of them if it is wanted.
# we can do the same for columns 'monthly_premium_auto' and 'total_claim_amount' 
```
